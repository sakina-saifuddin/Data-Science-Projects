#### vi. Handling dimensionality reduction, if needed

##### Steps to Apply PCA:

Standardize the Data:
PCA works better when all the features are on the same scale (e.g., if one feature has values in thousands and another in ones, PCA will be biased towards the larger numbers). So, we standardize the data first.

##### Apply PCA:
After standardizing, PCA will find the new features (principal components).

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler


# Step 1: Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(movies_copy[['Gross', 'Year']] )

from sklearn.preporcessing from StandardScaler
scaler=StandardScaler
sclaed=scaler.fit_transform(movies_copy[['Gross', 'Year']])

# Step 2: Apply PCA (reduce to 2 components)
pca = PCA(n_components=2)
print("pca", pca)
pca_result = pca.fit_transform(scaled_data)
print("pca_result", pca_result)

# Add the PCA results to the DataFrame
movies_copy['PCA1'] = pca_result[:, 0]
movies_copy['PCA2'] = pca_result[:, 1] 

# Display the result
print(movies_copy) # PCA is like creating new, smarter columns from your existing data.
 
```

**Observation**

- Principal Component Analysis (PCA) was applied to the Gross and Year columns after standardizing the data, reducing it to 2 components.
- The new PCA1 and PCA2 columns were created to represent the most important features (principal components) derived from the original data, allowing for more effective analysis.
